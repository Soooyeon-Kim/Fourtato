## Daily Percent Change with Matplotlib
### Matplotlib 을 이용한 일간 주가 변동 추세 확인

```python
Namyang  = pdr.get_data_yahoo('003920.KS',start='2016-01-01')
Ottogi = pdr.get_data_yahoo('007310.KS',start='2016-01-01')
```

```python
Namyang_dpc = (Namyang['Close']-Namyang['Close'].shift(1)) / Namyang['Close'].shift(1)*100
Namyang_dpc.iloc[0]=0
print(Namyang_dpc.describe()) #남양 주식 데이터 정보
```

```python
#일간변동추세가 가장 하락했던 날을 알아보기
Ottogi_dpc = (Ottogi['Close']-Ottogi['Close'].shift(1)) / Ottogi['Close'].shift(1)*100
Ottogi_dpc.iloc[0]=0
print(Ottogi_dpc.describe())
```

## Cumulative Sum 
### 누적 수익률 구하기

#### ----------------남양-----------------------

```python
print(Namyang_dpc.cumsum()) #오늘까지에 대한 전체 수익률
```

```python
#지난 3년간 일간 누적 변동률이 대략 -77% 인 것을 알 수 있다!!!!
```

```python
Namyang_dpc_dcs = Namyang_dpc.cumsum()
```

```python
plt.plot(Namyang.index,Namyang_dpc_dcs,'b',label='Namyang')
plt.ylabel('change %')
plt.grid(1)
plt.show()
```

#### --------------------오뚜기--------------------------------

```python
print(Ottogi_dpc.cumsum()) #오늘까지에 대한 전체 수익률
```

```python
Ottogi_dpc_dcs = Ottogi_dpc.cumsum()
```

```python
plt.plot(Ottogi.index,Ottogi_dpc_dcs,'b',label='Ottogi')
plt.ylabel('change %')
plt.grid(1)
plt.show()
```

```python
plt.title("Cumulative Sum of Ottogi and Namyan")
plt.plot(Ottogi.index, Ottogi_dpc_dcs, 'g', label='Ottogi')
plt.plot(Namyang.index, Namyang_dpc_dcs, 'b', label='Namyang')
plt.ylabel('Change %')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

## 남양의 일별시세 크롤링

#### df = df.sort_values(by='날짜') 
#### 를 입력하면  오름차순으로 정렬


```python
url = 'https://finance.naver.com/item/sise_day.nhn?code=003920&page=1'
html = BeautifulSoup(requests.get
    (url, headers={'User-agent': 'Mozilla/5.0'}).text, 'lxml')

with urlopen(url) as doc:
    pgrr = html.find('td', class_='pgRR')
    s = str(pgrr.a['href']).split('=')
    last_page = s[-1]

df = pd.DataFrame()
sise_url = 'https://finance.naver.com/item/sise_day.nhn?code=003920'  

for page in range(1, int(last_page)+1):
    page_url = '{}&page={}'.format(sise_url, page)
    df = df.append(pd.read_html(requests.get(page_url,
        headers={'User-agent': 'Mozilla/5.0'}).text)[0])
    
df = df.dropna()
print(df)
```

```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import mplfinance as mpf
import requests
```

## 남양의 30일 거래 차트

```python
df = df.iloc[0:30]
df = df.rename(columns={'날짜':'Date', '시가':'Open', '고가':'High', '저가':'Low', '종가':'Close', '거래량':'Volume'})
df = df.sort_values(by='Date')
df.index = pd.to_datetime(df.Date)
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]

mpf.plot(df, title='Namyang candle chart', type='candle')

mpf.plot(df, title='Namyang ohlc chart', type='ohlc')

kwargs = dict(title='Namyang customized chart', type='candle',
    mav=(2, 4, 6), volume=True, ylabel='ohlc candles')
mc = mpf.make_marketcolors(up='r', down='b', inherit=True)
s  = mpf.make_mpf_style(marketcolors=mc)
mpf.plot(df, **kwargs, style=s)

```
